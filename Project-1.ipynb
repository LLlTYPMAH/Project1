3. Следующим этапом преобразуем признак **"Опыт работы"**. Его текущий формат - это: **<Опыт работы: n лет m месяцев, периоды работы в различных компаниях…>**. 

Из столбца нам необходимо выделить общий опыт работы соискателя в месяцах, новый признак назовем "Опыт работы (месяц)"

Для начала обсудим условия решения задачи:
* Во-первых, в данном признаке есть пропуски. Условимся, что если мы встречаем пропуск, оставляем его как есть (функция-преобразование возвращает NaN)
* Во-вторых, в данном признаке есть скрытые пропуски. Для некоторых соискателей в столбце стоит значения "Не указано". Их тоже обозначим как NaN (функция-преобразование возвращает NaN)
* В-третьих, нас не интересует информация, которая описывается после указания опыта работы (периоды работы в различных компаниях)
* В-четвертых, у нас есть проблема: опыт работы может быть представлен только в годах или только месяцах. Например, можно встретить следующие варианты:
    * Опыт работы 3 года 2 месяца…
    * Опыт работы 4 года…
    * Опыт работы 11 месяцев…
    * Учитывайте эту особенность в вашем коде

Учитывайте эту особенность в вашем коде

В результате преобразования у вас должен получиться столбец, содержащий информацию о том, сколько месяцев проработал соискатель.
Выполните преобразование, ответьте на контрольные вопросы и удалите столбец **"Опыт работы"** из таблицы.


```python
# Функция принимет строку и возвращает число месяцев
def exp_tr(arg):
    # получение списка из первых 6 слов в строке (что бы случайно не найти совпадение в навании компании)
    arg = list(str(arg).split(' '))[0:6]
    a=0
    
    # Цикл по этому списку. В случае соответствия элемента и ключегого слова, прибавление счетчика на предыдущий элемент для месяца и на предыдущий элемент * 12 для года
    for i in range(len(arg)):
        if 'лет' in arg[i] or 'год' in arg[i]:
            a += int(arg[i-1])*12
        elif 'месяц' in arg[i]:
            a += int(arg[i-1])
            
    # Если строка пуста или там значение типа 'Не указанно', вернуть None
    if a == 0: return None
    return int(a)

# Применение функции для новой категории, отображение её медины
data['Опыт работы (месяц)'] = data['Опыт работы'].apply(exp_tr)
display(data['Опыт работы (месяц)'].median())
```

```python
# Список городов-миллинноков
million_cities = ['Новосибирск', 'Екатеринбург', 'Нижний Новгород','Казань',
                  'Челябинск', 'Омск', 'Самара', 'Ростов-на-Дону', 'Уфа',
                  'Красноярск', 'Пермь', 'Воронеж', 'Волгоград' ]

# Функция принимает строку и возвращает одну из 4х категорий
def city_extr(arg):
    # Отделение первого слова
    arg = list(arg.split(' '))[0]
    
    # Сравнение с ключевыми словами, сравнение со списком ключевых слов, категория 'другие' по остаточному признаку
    if arg == 'Москва' or arg == 'Санкт-Петербург': return arg
    elif arg in million_cities: return 'город-миллионник'
    else: return 'другие'
    
# Применение функции для новой категории, отображение соотношения ее значений
data['Город'] = data['Город, переезд, командировки'].apply(city_extr)
display(data['Город'].value_counts(True))


# Функция принимает строку и возвращает булиево значени
def reloc_extr(arg):
    # Получение списка значений
    arg = list(arg.split(' , '))
    
    # Необходимое значение стоит третьим, если указано метро, и вторым, если нет.
    if 'м.' in arg[1]: i = 2
    else: i = 1
    
    # Проверка, на присутсвие частицы 'не', если ее нет, значит готов
    if 'не' not in arg[i]: return True
    else: return False
    
data['Готовность к переезду'] = data['Город, переезд, командировки'].apply(reloc_extr)


def b_trp_extr(arg):
    # Получение списка значений
    arg = list(arg.split(' , '))
    
    # Необходимое значение стоит четвёртым, если указано метро, и третьим, если нет.
    # Проверка указания метро, проверка на наличие последнего значения(указанно, что его может не быть)
    if 'м.' in arg[1]:
        i = 3
        if len(arg) == 3: return False
    else:
        i = 2
        if len(arg) == 2: return False
        
    # Проверка, на присутсвие частицы 'не', если ее нет, значит готов
    if 'не' not in arg[i]: return True
    else: return False
    
data['Готовность к командировкам'] = data['Город, переезд, командировки'].apply(b_trp_extr)


data.head()
#display(data[data['Готовность к командировкам'] & data['Готовность к переезду']].count()/data.count())
```

5. Рассмотрим поближе признаки **"Занятость"** и **"График"**. Сейчас признаки представляют собой набор категорий желаемой занятости (полная занятость, частичная занятость, проектная работа, волонтерство, стажировка) и желаемого графика работы (полный день, сменный график, гибкий график, удаленная работа, вахтовый метод).
На сайте hh.ru соискатель может указывать различные комбинации данных категорий, например:
* полная занятость, частичная занятость
* частичная занятость, проектная работа, волонтерство
* полный день, удаленная работа
* вахтовый метод, гибкий график, удаленная работа, полная занятость

Такой вариант признаков имеет множество различных комбинаций, а значит множество уникальных значений, что мешает анализу. Нужно это исправить!

Давайте создадим признаки-мигалки для каждой категории: если категория присутствует в списке желаемых соискателем, то в столбце на месте строки рассматриваемого соискателя ставится True, иначе - False.

Такой метод преобразования категориальных признаков называется One Hot Encoding и его схема представлена на рисунке ниже:
<img src=https://raw.githubusercontent.com/AndreyRysistov/DatasetsForPandas/main/ohe.jpg>
Выполните данное преобразование для признаков "Занятость" и "График", ответьте на контрольные вопросы, после чего удалите их из таблицы

```python
# Список возможных значений в графе 'Занятость'
arr1 = ['полная занятость', 'частичная занятость', 'проектная работа', 'волонтерство', 'стажировка']


# Цикл по значениям создает новую графу с этим значением и заполняет его булиевой переменной для строки, в зависимости от наличия этого значения в проверяемой строке
for i in arr1:
    data[i] = data['Занятость'].apply(lambda arg: True if i in arg else False)

# Список возможных значений в графе 'График'
arr2 = ['полный день', 'сменный график', 'гибкий график', 'удалённая работа', 'вахтовый метод']

# Цикл по значениям создает новую графу с этим значением и заполняет его булиевой переменной для строки, в зависимости от наличия этого значения в проверяемой строке
for i in arr2:
    data[i] = data['График'].apply(lambda arg: True if i in arg else False)

# Отображение когличества требуемых условий в полученных графах
display(data[data['проектная работа'] & data['волонтерство']].shape[0])
display(data[data['вахтовый метод'] & data['гибкий график']].shape[0])
```

4. Умножить сумму желаемой заработной платы на присоединенный курс валюты (close) и разделить на пропорцию (обратите внимание на пропуски после объединения в этих столбцах), результат занести в новый столбец "ЗП (руб)".


```python
# Получение и преобразование датафрейма с курсами валют
rate = pd.read_csv('data/ExchangeRates.csv', sep=',', usecols=['currency', 'date', 'proportion', 'close'])
rate['date'] = pd.to_datetime(rate['date'], dayfirst=True)

# Немного костыльное отделение даты в исходной таблице
data['Обновление резюме'] = pd.to_datetime(data['Обновление резюме'])
data['date'] = pd.to_datetime(data['Обновление резюме'].dt.date)

# Словарь соотвествий наименований валют в исходной таблице и таблице с курсами соответственно
dic = {
    'грн':'UAH',
    'USD':'USD',
    'EUR':'EUR',
    'бел.руб':'BYN',
    'KGS':'KGS',
    'сум':'UZS',
    'AZN':'AZN',
    'KZT':'KZT',
    'руб':'RUB'
}

# Функция принимет строку и возвращает соответсвующее значение из словаря dic
def curr(arg):
    # Получение второго значения в строке
    arg = str(arg.split(' ')[1])
    # Удаление точки в конце
    if arg[-1] =='.': arg = arg[:-1]
    
    # Подстановка наименования из словаря по ключу
    if arg in dic:
        return dic[arg]
    else:
        return None

# Создание столбца наименований валют
data['currency'] = data['ЗП'].apply(curr)

# Не окончательная версия столбца 'ЗП(руб)' формируется из первого значения в 'ЗП' переведенного в число. Перевод в рубли по курсу далее
data['ЗП(руб)'] = data['ЗП'].apply(lambda arg: int(arg.split(' ')[0]))

# Обединение таблиц по дате и наменованиям валют для получения соответсвий с актуальным курсом
data = data.merge(
    rate,
    on=['date', 'currency'],
    how='left')

# В датафрейме с курсами валют не было значений для рубля, поэтому на их месте остались пропуски
# Заполнение пропусков еденицей для корректных вычислений
data['close'] = data['close'].fillna(1)
data['proportion'] = data['proportion'].fillna(1)

# Окончательная версия столбца 'ЗП(руб)' с переводом в рубли
data['ЗП(руб)'] = (data['ЗП(руб)'] * data['close']) / data['proportion']

```

```python
# Удаление всех ненужных столбцов
data.drop(['ЗП', 'close', 'proportion', 'date', 'currency', 'Опыт работы'], axis=1, inplace=True)
display(data['ЗП(руб)'].median())
```

```python
# Функция нахождения выбросов по методу z-отклонений.
# Получает на вход Таблицу, Рассматриваемый признак, Использовать ли логарифмический масштаб и Количетво сигм слева и справа
# Возвращает выбросы и очещенную таблицу
def outliers_z_score_mod(data, feature, log_scale=False, left = 3, right = 3):
    # Логарифирование значений признака, если это указанно
    if log_scale:
        x = np.log(data[feature]+1)
    else:
        x = data[feature]
    # Мю - среднее по столбцу
    mu = x.mean()
    # Сигма - стандартное отклонение
    sigma = x.std()
    
    # Вычисление нижней и вержней границы по методу
    lower_bound = mu - left * sigma
    upper_bound = mu + right * sigma
    
    # В выбросы попадает всё, что снаружи границ, в очищенную таблицу, всё, что внутри
    outliers = data[(x < lower_bound) | (x > upper_bound)]
    cleaned = data[(x >= lower_bound) & (x <= upper_bound)]
    return outliers, cleaned

# Отображение графика распределения возраста в логарифмическом масштабе
fig, axes = plt.subplots(figsize=(10, 8))
log_age= np.log(data['Возраст']+1)
histplot = sns.histplot(data=log_age, bins=30)

# Отображающие линий среднего и границ интервала метода трех сигм (без послабления)
histplot.axvline(log_age.mean(), color='k', lw=2)
histplot.axvline(log_age.mean()+ 3 * log_age.std(), color='k', ls='--', lw=2)
histplot.axvline(log_age.mean()- 3 * log_age.std(), color='k', ls='--', lw=2)

# Добавление названия и подписей осей
axes.set_title('Логарифмическое распределение возраста', fontsize=16)
axes.set_ylabel('Количество', fontsize=12) 
axes.set_xlabel('Возраст', fontsize=12)

# Отображение выбросов по методу z-отклонений с послабдением на одну сигму вправо
display(outliers_z_score_mod(data, 'Возраст', log_scale=True, left = 3, right = 4)[0])

# Очистка таблицы по методу z-отклонений с послабдением на одну сигму вправо
data = outliers_z_score_mod(data, 'Возраст', log_scale=True, left = 3, right = 4)[1]

## гыгыгы ООО ЖМЫХ
```
